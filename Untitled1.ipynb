### **3 - Qual o tipo de distribuição das variáveis?** 

Para identificar a distribuição das variáveis, irei plotar o histograma. Sobre o histograma podemos dizer que são representações gráficas de dados que mostram a distribuição de valores em um conjunto de dados. Eles são comumente usados em estatística para ilustrar a distribuição de frequência de uma variável contínua ou discreta. São úteis para identificar a moda (valor mais frequente), mediana (valor central) e a forma geral da distribuição de dados. Eles também podem ser usados para identificar valores atípicos ou extremos no conjunto de dados.

```python
# plotar histograma das variáveis numéricas
df.hist(bins=15, figsize=(20,12));
```

###**4 - Há *outliers* presentes?**

Outliers, valores atípicos, são pontos de dados que se afastam significativamente do padrão ou da tendência geral de um conjunto de dados. Esses valores são geralmente muito maiores ou menores do que a maioria dos outros valores do conjunto de dados e podem ser causados por erros de medição, erros de entrada de dados ou por fenômenos naturais incomuns.
Pela distribuição do histograma, é possível verificar indícios da presença de *outliers*. Olhe por exemplo a variável`calculated_host_listings_count`.

Os valores não seguem uma distruição, e distorcem toda a representação gráfica. Para confirmar, há duas maneiras rápidas que auxiliam a detecção de *outliers*. São elas:

* Resumo estatístico por meio do método `describe()`
* Plotar `boxplots` para a variável.



```python
# ver o resumo estatístico das variáveis numéricas
df[['price','minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']].describe()
```

Pelo resumo estatístico acima, podemos confirmar algumas hipóteses apresentadas como na variável price em que a média está em 177,11 dólares mas que o valor de 211,25 representa 75%  dos imóveis.


```python
# minimun nights
df.minimum_nights.plot(kind='box', vert=False, figsize=(15,3))
plt.show()

# ver quantidade de valores acima de 30 dias para minimum_nights
print('minimum_nights: valores acima de 30:')
print('{} entradas'.format(df[df.minimum_nights > 30]))
print('{:.4f}%'.format((len(df[df.minimum_nights > 30]) / df.shape[0])*100))
```

#### Boxplot para price

```python
# price
df.price.plot(kind='box', vert=False, figsize=(20, 2),) 
plt.show()

# ver quantidade de valolres acima de 1500 para price
print("\nprice: valores acima de 1500")
print("{} entradas".format(len(df[df.price >1500])))
print('{:.4f}%'.format((len(df[df.price > 1500]) / df.shape[0])*100))

# df.price. plot(kind='box', vert=False, xlim=(0,1300),figsze=(15,3));
```

#### Histogramas sem *outliers*

Já que identificamos *outliers* nas variáveis `price` e `minimum_nights`, vamos agora limpar o *DataFrame* delas e plotar novamente o histograma.

```python
# remover os *outliers* em um novo DataFrame
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 1500].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 30].index, axis=0, inplace=True)

# remover `neighbourhood_group`, pois está vazio
df_clean.drop('neighbourhood_group', axis=1, inplace=True)

# plotar o histograma para as variáveis numéricas
df_clean.hist(bins=15, figsize=(15,10));
```

### **5 - Qual a correlação existente entre as variáveis**

A matriz de correlação significa que existe uma relação entre duas coisas. No nosso contexto, estamos buscando relação ou semelhança entre duas variáveis.

Essa relação pode ser medida, e é função do coeficiente de correlação estabelecer qual a intensidade dela. Para valores próximos de 1, indicam uma correlação positiva forte e para valores próximos de -1, uma correlação negativa forte. Para identificar as correlações existentes entre as variávfeis de interresse vou: 

* Criar uma matriz de correlação
* Gerar um *heatmap* a partir dessa matriz, usando a biblioteca `seaborn`

```python
# criar uma matriz de correlação
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()

display(corr)
```

```python
sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True)
```

### **6 -  Qual o tipo de imóvel mais alugado no Airbnb?**

A coluna da variável `room_type` indica o tipo de locação que está anunciada no Airbnb. Se você já alugou no site, sabe que existem opções de apartamentos/casas inteiras, apenas o aluguel de um quarto ou mesmo dividir o quarto com outras pessoas.

Vamos contar a quantidade de ocorrências de cada tipo de aluguel, usando o método `value_counts()`.

```python
# mostrar a quantidade de cada tipo de imóvel disponível
df_clean.room_type.value_counts()
```

```python
# mostrar a porcentagem de cada tipo de imóvel disponível
df_clean.room_type.value_counts()/df_clean.shape[0] 
```

Dados os valores apresentados acima, private room representa 52% das acomodações, entire home/apt 47%, enquanto shared room uma porcentagem bem insignificável.

### **7 - Qual a localidade mais cara de Cambridge?**

Uma maneira de se verificar uma variável em função da outra é usando `groupby()`. No caso, queremos comparar os bairros (*neighbourhoods*) a partir do preço de locação.

```python
df_clean.groupby(['neighbourhood']).price.mean().sort_values(ascending=False)[:10]
```

Nos dados acima vemos que o bairro de Cambridge Highlands possui uma relevância maior em relação aos outros bairros, com o valor de 336 dólares de média, West Cmabridge e East Cambridge vem na seguência com os valores em me´dia de 206 e 200 dólares. 
Para dar um exemplo de como a amostra pode não ser representativa apenas se compararmos o preço médio, sempre devemos analisar o máximo de dados possíveis, vamos ver quantas entradas há para Cambridge Highlights.

```python
# ver a quantidade de imóveis em Cambridge Highligths
print(df_clean[df_clean.neighbourhood == "Cambridge Highlights"].shape)

# ver as 5 entradas do Cambridge Highligths
df_clean[df_clean.neighbourhood=='Cambridge Highligths']
```

```python
# ver a quantidade de imóveis em Riverside
print(df_clean[df_clean.neighbourhood=='Riverside'].shape)

# ver as entradas do Riverside
df_clean[df_clean.neighbourhood=='Riverside']
```

Acima vemos que apesar da média de preço ser mais elevada em relação as outras localidades, Cambrdge Highlights nõa possui imóvel disponível para aluguel e que em Riverside apesar da média mais baixa em relação ao citado acima, há vários imóveis disponíveis.
Como são fornecidas as Latitudes e Longitudes para os imóveis, é possível plotar cada ponto. Para isso, considera-se `x=longitude` e `y=latitude`.

```python
# plotar os imóveis pela latitude-longitude
df_clean.plot(kind="scatter", x='longitude', y='latitude', alpha=0.4, c=df_clean['price'], s=8,
              cmap=plt.get_cmap('jet'), figsize=(12,8));
```
